# Machine Learning Model Fitting

#### Lets first split the data set into a training and testing group

```python
X = df.drop('stiffness_value', axis=1) # Set the X values as all the feature entries
y = df['stiffness_value'] # Set the y values to all the stiffness values


# # Notes on test-train split:
# # - 25% of the data will be used for testing at the end
# # - Random state of 0 has been used for reproducibility of results

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

## Linear Regression Model: Stochastic Gradient Descent (SGD) Regressor

```python
# Define the hyperparameter space for bayesian identification
search_space = {
    'sgd__alpha': Real(1e-6, 1e-1, prior='log-uniform'),
    'sgd__loss': Categorical(['squared_error', 'huber', 'epsilon_insensitive', 'squared_epsilon_insensitive']),
    'sgd__penalty': Categorical(['none', 'l2', 'l1', 'elasticnet']),
    'sgd__l1_ratio': Real(0, 1),
    'sgd__learning_rate': Categorical(['constant', 'optimal', 'invscaling', 'adaptive']),
    'sgd__eta0': Real(1e-5, 1e-2, prior='log-uniform')  # Only relevant if learning_rate is 'constant' or 'invscaling'
}

# Define the SGDRegressor within a pipeline so that data is regularized before hand
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('sgd', SGDRegressor(max_iter=1000, tol=1e-3))
])

# Use Baysian Identification via bayes rule
bayes_search = BayesSearchCV(pipeline, search_space, n_iter=32, scoring='neg_mean_squared_error', cv=5) # Cross-validation done here

# Fit the model with the optimizer
bayes_search.fit(X_train, y_train)
```

```python
# Get the best parameters and the corresponding score

best_params = bayes_search.best_params_
best_score = bayes_search.best_score_

# print("Cross-Validation Results:", bayes_search.cv_results_)
print("Best parameters found: ", best_params)
print("Best cross-validated score (negative MSE): ", best_score)
```

```python
# Extract the best model retrained on the whole training dataset
model_linear = bayes_search.best_estimator_
```

```python
# Now lets test this model against the testing set to see how it does!

# Use the best estimator to make predictions on the test set
y_pred = model_linear.predict(X_test)

# Calculate performance metrics
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'Root Mean Squared Error: {rmse}')
print(f'R-squared score: {r2}')
```

### Linear: Visualization of performance

```python
# Plot actual vs. predicted values
# Actual vs. Predicted Values Plot
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.5, label='Predicted vs Actual')
# Diagonal line represents perfect predictions
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2, label='Perfect predictions')
plt.xlabel('Actual Stiffness')
plt.ylabel('Predicted Stiffness')
plt.title('Linear Regression: Actual vs. Predicted Values')
plt.legend()  # This adds the legend to the plot
plt.show()
```

```python
# Residuals Plot
residuals = y_test - y_pred
plt.figure(figsize=(10, 6))
sns.histplot(residuals, kde=True)
plt.xlabel('Residual')
plt.ylabel('Frequency')
plt.title('Linear Regression: Distribution of Residuals')
plt.axvline(x=0, color='k', linestyle='--', label='Zero Error')  # Adds a line for zero error
plt.legend()
plt.show()
```

## Gaussian Process Regression (GPR) Model
- Use RBF and White Noise Kernel as kernel

```python
# Define the kernel
kernel = ConstantKernel() * RBF() + WhiteKernel()

# Print all available parameter names for this kernel
print(kernel.get_params().keys())
```

```python
# Create a pipeline that first standardizes the data, then applies Gaussian Process Regressor
gp_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('gp', GaussianProcessRegressor(kernel=RBF(length_scale=10, length_scale_bounds=(1e-2, 1e2))))
])

# Define the hyperparameter search space
search_spaces = {
    'gp__kernel__length_scale': Real(1e-2, 1e2, prior='log-uniform')
}

# Set up the BayesSearchCV object with 5-fold cross-validation
bayes_search = BayesSearchCV(estimator=gp_pipeline, search_spaces=search_spaces, n_iter=15, cv=5)

# Fit the BayesSearchCV object to the standardized data
bayes_search.fit(X_train, y_train)
```

```python
# Extract the best model and its parameters
model_gp = bayes_search.best_estimator_
best_hyperparams = bayes_search.best_params_
best_score = bayes_search.best_score_

print(f"Best hyperparameters: {best_hyperparams}")
print(f"Best cross-validated score: {best_score}")
```

```python
# Now lets test this model against the testing set to see how it does!

# Use the best estimator to make predictions on the test set
y_pred = model_gp.predict(X_test)

# Calculate performance metrics
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'Root Mean Squared Error: {rmse}')
print(f'R-squared score: {r2}')
```

## Artificial Neural Network (ANN) Model
- Optimizer: Adam Optimizer
- Activation Function: Gaussian Error Linear Unit (GeLU)
- Loss Function: Mean Squared Error (MSE)
- Training Time: 100 Epochs
- Validation Split: 20% for validation

```python
# Begin by regularizing our inputs and outputs

# Standardize the input features
scaler_X = StandardScaler()
X_train_scaled = scaler_X.fit_transform(X_train)
X_test_scaled = scaler_X.transform(X_test)

# Standardize the target values
scaler_y = StandardScaler()
y_train_scaled = scaler_y.fit_transform(y_train.values.reshape(-1, 1)).flatten()
y_test_scaled = scaler_y.transform(y_test.values.reshape(-1, 1)).flatten()
```

#### Model 1:
- One Hidden Layer: 16 Nodes

```python
# Define the neural network model with GeLU activation

# Lets look at one hidden layer first with 16 nodes

model = Sequential([
    Dense(16, activation=tf.keras.activations.gelu, input_shape=(X_train_scaled.shape[1],)),
    Dense(1)
])
```

```python
# Compile the model
model.compile(optimizer='adam', loss='mse')

# Train the model
history = model.fit(X_train_scaled, y_train_scaled, epochs=100, validation_split=0.2, verbose=1)
```

```python
# Evaluate the model on the test data
test_loss = model.evaluate(X_test_scaled, y_test_scaled)
print(f'Test MSE: {test_loss}')
```

```python
# Plot the training and validation MSE at each epoch
plt.plot(history.history['loss'], label='Training MSE')
plt.plot(history.history['val_loss'], label='Validation MSE')
plt.xlabel('Epoch')
plt.ylabel('Mean Squared Error')
plt.title('Model 1 (1 Layer: 16 Nodes) Performance (MSE) at Each Epoch')
plt.legend()
plt.show()
```

#### Model 7:
- Two Hidden Layers: 32 Nodes --> 16 Nodes (Best Model so far)
- Applying several features to try and combat overfitting
    - L2 Regularizer
    - Dropout layers
    - Batch Normalization added after droupout layers
    - Early stopping condition added when validation loss stops improving

```python
# Define the neural network model with dropout and L2 regularization
model = Sequential([
    Dense(64, activation='gelu', kernel_regularizer=l2(0.001), input_shape=(X_train_scaled.shape[1],)),
    Dropout(0.5),
    BatchNormalization(),
    Dense(64, activation='gelu', kernel_regularizer=l2(0.001)),
    Dropout(0.5),
    BatchNormalization(),
    Dense(1)
])
```

```python
# Compile the model
model.compile(optimizer='adam', loss='mse')

# Define an EarlyStopping callback
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
# Train the model with the EarlyStopping callback
history = model.fit(X_train_scaled, y_train_scaled, epochs=100, validation_split=0.2, callbacks=[early_stopping], verbose=1)
```

```python
# Evaluate the model on the test data
test_loss = model.evaluate(X_test_scaled, y_test_scaled)
print(f'Test MSE: {test_loss}')
```

#### This doesn't have better performance than prior models but maybe can help alleviate over fit?
